```python
text = []

with open("day4.txt", "r") as file:
  for line in file: text.append(line.strip())
```

```python
"""
Find all occurances of XMAS in a matrix of text. Occurances can be horizontal,
vertical, diagonal, backwards, and overlapping
"""
def examine(text: list[str]) -> int:
  total = 0

  for i in range(len(text)):
    for j in range(len(text[i])):

      # horizontal cases
      if j < len(text[i]) - 3:
        x = text[i][j:j+4]
        if x == "XMAS" or x == "SAMX":
          total += 1

      # vertical cases
      if i < len(text) - 3:
        x = text[i][j] + text[i+1][j] + text[i+2][j] + text[i+3][j]
        if x == "XMAS" or x == "SAMX":
          total += 1

      # negative diagonal
      if i < len(text) - 3 and j < len(text[i]) - 3:
        x = text[i][j] + text[i+1][j+1] + text[i+2][j+2] + text[i+3][j+3]
        if x == "XMAS" or x == "SAMX":
          total += 1

      # positive diagonal
      if i >= 3 and j < len(text[i]) - 3:
        x = text[i][j] + text[i-1][j+1] + text[i-2][j+2] + text[i-3][j+3]
        if x == "XMAS" or x == "SAMX":
          total += 1
  return total

x = examine(text)
print(x)
```

```python
"""
Find all x-mas in a text matrix. i.e. two occurances of "MAS" in an X pattern
example occurance:
     M.M
     .A.
     S.S
Return: the total count of occurances of the X pattern.
"""
def x_amine(text: list[str]) -> int:
  total = 0

  for i in range(len(text)):
    for j in range(len(text[i])):
      if 0 < j < len(text) - 1 and 0 < i < len(text[i]) - 1:
        if text[i][j] == "A":
          if text[i-1][j-1] == "M" and text[i-1][j+1] == "M" and \
             text[i+1][j-1] == "S" and text[i+1][j+1] == "S":
            total += 1

          if text[i-1][j-1] == "M" and text[i-1][j+1] == "S" and \
             text[i+1][j-1] == "M" and text[i+1][j+1] == "S":
            total += 1

          if text[i-1][j-1] == "S" and text[i-1][j+1] == "S" and \
             text[i+1][j-1] == "M" and text[i+1][j+1] == "M":
            total += 1

          if text[i-1][j-1] == "S" and text[i-1][j+1] == "M" and \
             text[i+1][j-1] == "S" and text[i+1][j+1] == "M":
            total += 1

  return total

x = x_amine(text)
print(x)
```
